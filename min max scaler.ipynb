```python
# we have to pick up tyhe some useful columns only
df=pd.read_csv("E:\\6th semmmm\\feature engineering datasets\\min max\\wine_data.csv",
               header=None,usecols=[0,1,2])
df.head()
```

```python
df.columns=['class label','Alcohol','Malic Acid']
df.head()
```

```python
# vwe have to perform the train test spliy
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(df.drop('class label' ,axis=1),
                                                 df['class label'] , test_size=0.3,random_state=10)
print("split the data sucessfully")
```

# min max scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

# fit the scaler to train set , it will learn the parameters like max and the mean 
scaler.fit(x_train)

# btransform train and test sets 
x_train_scaled=scaler.transform(x_train)  # numpy array 
x_test_scaled=scaler.transform(x_test)  # numpy array
print(" sucess")
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
# convert the numpy array into the data frames 
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
print(" changed sucessfully")
```

# perform the logistic regfression before and the after scalings 
# and check the accuracy testings

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr_scaled=LogisticRegression()

lr.fit(x_train,y_train)
lr_scaled.fit(x_train_scaled,y_train)
print("fit sucessfully")
```

```python
# predictions 
y_pred=lr.predict(x_test)
y_pred_scaled=lr_scaled.predict(x_test_scaled)
```

```python
y_pred  # numpy arrays 
```

```python
# check the accuracy
from sklearn.metrics import accuracy_score
print(" actual ",accuracy_score(y_test,y_pred))
print("scaled" ,accuracy_score(y_test,y_pred_scaled))
```
